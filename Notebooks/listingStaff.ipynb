# 1. Dictionnary <a id='Dictionnary'></a>

We first assign a dictionnary called `staff` that will provide for each teaching staff the following information:

+ name  
+ position within the School
+ units: a list of units taught by the chosen academic  
+ perc: the percentage of the unit taught by the chosen academic 

````python
staff = {}

staff['name'] = {}
staff['position'] = {} 
staff['units'] = {} 
staff['perc'] = {} 
````

<div class="alert alert-block alert-danger">
It is worth noting that units and perc have to be listed in the same order.
</div>

We also have to iterate the `k` variable when adding new staff.

As an example to add a new teaching staff do the following:
 
```python
k += 1
staff['name'][k] = 'Newton'
staff['position'][k] = 'Academic'
staff['units'][k] = ['PHYS1727']
staff['perc'][k] = [50.]
```
