# 1. Dictionnary <a id='Dictionnary'></a>

<div class="alert alert-block alert-danger">
It is worth noting that the load needs to reflect how hard a unit is to teach...
</div>

+ code
+ level: Level corresponds to first (=0), second (=1), third (=2) year and master (=3)
+ description
+ coordinator
+ load: a standard 13 weeks unit will correspond to a load of 100.
